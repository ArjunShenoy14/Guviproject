```python

import re
regex = re.compile(r'([A-Za-z0-9]+[.-_])*[A-Za-z0-9]+@[A-Za-z0-9-]+(\.[A-Z|a-z]{2,})+')

#sign up function
def signup():
  y = True
  x = True

  email=input("Enter email address")
  if re.fullmatch(regex, email):
    y= False
    print("Valid Email")
    #valid email
  else:
    y = True
    #invalid email
    
  p=input("Enter password:")

  while x:  
    if (len(p)<5 or len(p)>16):
        break
    elif not re.search("[a-z]",p):
        break
    elif not re.search("[0-9]",p):
        break
    elif not re.search("[A-Z]",p):
        break
    elif not re.search("[$#@]",p):
        break
    elif re.search("\s",p):
        break
    else:
        print("Valid Password")
        x=False
        break

  if x == False and y == False:
    with open("credentials.txt","a") as f:
      f.write(email+"\t")
      f.write(p+"\n")
    f.close()
    print("You have registered successfully!")
  else:
    print("Invalid Email or Password")

#login function
def login():
  login={}
  flag=0
  with open("credentials.txt","r") as f:
    for line in f:
      key,val= line.split("\t")
      #print(key+" and "+val)
      login[key]=val
      #print(login)
    f.close()
  em=input("Enter Email: ")
  ps=input("Enter Password: ")
  if em in login.keys():
    #print(login[em])
    pas=str(login[em])
    pas=pas.strip("\n")
    if ps == pas:
      flag=1
      #print("yes")
    else:
      flag=0
      #print("no")
  if (flag==0):
    print("Invalid Login Credentials \n1.Forgot password?\n2.Go Back")
    c=int(input("Enter choice: "))
    if c==1:
      resetpass(em)
    elif c==2:
     return
    else:
      print("invalid input")
  else:
    print("Login successful")

#reset password function
def resetpass(email):
  login={}
  em=email
  with open("credentials.txt","r") as f:
    for line in f:
      key,val= line.split("\t")
      #print(key+" and "+val)
      login[key]=val
      #print(login)
  new_pass=input("Enter new password: ")
  login[em]=new_pass
  with open("credentials.txt","w") as f:
    for key, value in login.items():
      f.write(key+"\t")
      f.write(value+"\n")
    f.close()
  print("Pasword changed successfully!")



#Main 
while 1:
  print("********** Login System **********")
  print("1.Sign up")
  print("2.Login")
  print("3.Exit")
  ch=int(input("Enter choice: "))
  if ch == 1:
    signup()
  elif ch==2:
    login()
  elif ch==3:
    break
  else:
    print("Wrong Choice")
```
